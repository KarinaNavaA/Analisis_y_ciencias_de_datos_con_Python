#### La función range()
range()  es un generador de índice.

```python
range(1, 11)
```

Para visualizar a los elementos de un generador, necesitamos bucles for

```python
for i in range(1,11):
  print(i)
```

```python
# Vamos a crear una lista por comprensión
L2 = [i for i in range(11)]
L2
```

```python
L2[0:7] #También se pueden hacer slices sobre listas
```

## Condicionales
